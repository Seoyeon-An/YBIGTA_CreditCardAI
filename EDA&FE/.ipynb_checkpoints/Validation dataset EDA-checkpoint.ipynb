## Class 피처와 상관관계가 높은 V3, V7, V10, V12, V14, V17 에 대하여 Outlier 삭제
(개인적인 생각으로 30개의 모든 피처변수를 사용하기에는 변수의 개수가 너무 많을것 같기도 하고, 추후에 pca 등의 차원축소를 통해서도 변수를 줄일 것 같다는 판단이 들어 일단 상관관계 값을 기준으로 유의미한 변수들만 선정해서 데이터처리를 진행해봤습니다.

```python
def get_outlier(df=None, column=None, weight=1.5):
    quantile_25 = np.percentile(df[column].values, 25)
    quantile_75 = np.percentile(df[column].values, 75)
    iqr = quantile_75 - quantile_25
    iqr_weight = iqr * weight
    lowest_val = quantile_25 - iqr_weight
    highest_val = quantile_75 + iqr_weight
    outlier_index = df[column][(df[column] < lowest_val) | (df[column] > highest_val)].index
    return outlier_index
```

```python
imp_feature_list = ['V3', 'V7', 'V10', 'V12', 'V14', 'V17']
for i in imp_feature_list:
    outlier_index = get_outlier(df=valid_df, column=i, weight=1.5)
    print('{} 피처 outlier 데이터 index:'.format(i), outlier_index )
```

## Outlier index들의 합집합인 리스트 생성 후 해당 index row 제거

```python
index_list = []
imp_feature_list = ['V3', 'V7', 'V10', 'V12', 'V14', 'V17']
for i in imp_feature_list:
    outlier_index = get_outlier(df=valid_df, column=i, weight=1.5)
    index_list += outlier_index.tolist()

index_list = list(set(index_list))
len(index_list)
valid_df.drop(index_list, axis = 0, inplace = True)
valid_df.head()
```

## 각 피처변수 정규화
(표준화는 필요시 진행하면 될 것 같습니다. 일단은 각 변수들의 최대, 최소값의 폭이 다양한 것을 먼저 해결해야 할 것 같다고 판단했습니다.)

```python
def get_minmaxscaled_df(df=None, column=None):
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(df[column].values.reshape(-1, 1))
    df[column] = scaled_data
    return df
```

```python
for i in valid_df.columns.tolist()[:-1]:
    get_minmaxscaled_df(df=valid_df, column=i)

valid_df.describe()
```
